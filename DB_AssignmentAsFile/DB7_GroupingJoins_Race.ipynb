##### Assignment 7: Grouping  and more Joins

_1. Using a single join and a grouping, write a query that allows you to find the `forename, surname, driverid` and the total number of wins for each driver in the `f1db` schema._

```python
%%sql SELECT forename, surname, driverid, SUM(results.rank) as totalofwins FROM drivers
JOIN results USING (driverid) WHERE results.rank = 1
GROUP BY driverid
ORDER BY totalofwins DESC;
```

> _There are **31 drivers** won in the different races, and the top driver won a total of **41 winnings( 8 - Kimi Räikkönen )**._

_2. Write a query that joins the tables `results, constructors` and `drivers` to show the amount of times a driver has driven a car from a constructor. In other words how many times one driver (for instance Schumacher) has driven a constructor (for instance Mercedes). Your table should have three fields: `drivers.driverref, constructors.name` and `count(*)` (showing the number of times the driver has driven the constructor)._

    A. Hint: You can group on more than one value

```python
%%sql SELECT drivers.driverref as driver,constructors.name as constructor,count(constructorid) as "car driven" FROM drivers
JOIN results USING (driverid)
JOIN constructors USING (constructorid)
GROUP BY driver, constructor
ORDER BY "car driven" DESC
LIMIT 25;
```

> _The result is limited by 25 since there is nothing indicated in the assignment, but just the show the top most cars that has been driven most often(count-constructorids) by specified drivers(driverref)._

_3. Now we have found the most driven vehicles even for those who didn't finish the race. Extend the query from 2 by removing all tuples from the result where the status is not "Finished"._

```python
%%sql SELECT drivers.driverref as driver,constructors.name as constructor,count(constructorid) as "car driven" FROM drivers
JOIN results USING (driverid)
JOIN constructors USING (constructorid)
JOIN status USING (statusid) WHERE status = 'Finished'
GROUP BY driver, constructor
ORDER BY "car driven" DESC
LIMIT 25;
```

> _The query has just added an additional condition of getting only those who finished the race by joining the table of status and check its value. Now, we can see that the highest count in `cars driven` remains the same but the lowest has only 44 times car driven(in #2 query, it was 91 times), meaning the tuples for drivers who did not finished the race were eliminated in the list._

_4. Create a third (and new) query that finds the amount of milliseconds spent in pitstop (see the `pitstops` relation) by each unique combinations of `driverid` and `raceid`._

    A. Hint: use the sum aggregate function to find the total amount of milliseconds
    B. Hint: i recommend making this into a view


```python
%%sql SELECT driverid,raceid, SUM(milliseconds) as ms FROM pitstops
GROUP BY driverid,raceid
ORDER BY raceid ASC
LIMIT 30;
```

```python
%%sql CREATE VIEW pitstops_view as
SELECT driverid,raceid, SUM(milliseconds) as ms FROM pitstops
GROUP BY driverid,raceid
ORDER BY raceid ASC;
```

> _New query for pitstops with the amount of milliseconds spent  is created with a view. As you can see, the tuples has been uniquely identified based on the combination of driverid and raceid._

_5. Time to put everything together. Use the query for 4 as a join subquery in query 3 (see slides on "a query within a query") to find the total amount of pitstop time, for each result in query 3_

```python
%%sql SELECT drivers.driverref as driver, constructors.name as car, count(constructorid) as cardriven, sum(ms) as totalpitstop  from drivers
JOIN results USING (driverid) 
JOIN pitstops_view USING (driverid, raceid)
JOIN constructors USING (constructorid)
JOIN status USING (statusid) WHERE status = 'Finished' 
GROUP BY drivers.driverref, constructors.name
ORDER BY cardriven DESC;
```

_6. As a last thing we want to find the driver with the least pistop time. However we want to control for the amount of races (see explanation below), so we need to include one last column: the average pitstop time per race. Your table should be sorted based on the ascending values of that average pitstop time._

    A. Explanation: If we just count the pitstop time, we only get the total time he spend in a workshop, no matter how many races he participated in. If a driver participated in 100 races, he would have more pitstop time than a driver participating in one race!
    B. Hint: Use the sum function to accumulate the total pistop time and divide that with the amount of races the driver has participated in


```python
%%sql SELECT drivers.driverref as driver, constructors.name as car,count(constructorid) as cardriven, sum(ms) as totalpitstop,ROUND(sum(ms)/count(results.raceid),0) as avgpitstop from drivers
JOIN results USING (driverid) 
JOIN pitstops_view USING (driverid, raceid)
JOIN constructors USING (constructorid)
JOIN status USING (statusid) WHERE status = 'Finished' 
GROUP BY drivers.driverref, constructors.name
ORDER BY avgpitstop ASC
LIMIT 10;
```

> _In the originally 76 rows returned by the query, I only took the 10 with the least average pitstop time that has been asked in the assignment._
